# Introduction to the Open-FF project

For the analysis in this paper, we used a data set from the Open-FF project (Allison, 2020), an effort to transform the collection of FracFocus disclosures into a research-quality data set.  The code and the resulting data sets are maintained at CodeOcean.com, a service that guarantees that published code is reproducible and that allows any user to inspect, modify and re-run the code.   Occasional new versions of the code add features and incorporate the most recent FracFocus published data.  The Open-FF version used for the analysis in this paper is version 7 which uses FracFocus data downloaded on July 31, 2020.

The data sets produced by Open-FF include the original, unmodified FracFocus fields coupled with generated fields that correct or add information to those raw fields. Users can, thus, see both the original “raw” data as well as the “value-added” data. Additionally, flags are included that indicate what types of issues have been found and allow end-users to easily filter the data based on the Open-FF standardization and quality control results.  Field names that begin with an upper-case letter mark the fields from the original FracFocus download; names beginning with a lower-case letter are the fields generated by Open-FF.  The CodeOcean site has a data dictionary that describes each generated field.
